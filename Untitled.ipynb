```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# load variables from other notebook (Ontario_Covid_19_Graphs.ipynb)
%store -r data
%store -r daily_cases
%store -r daily_recoveries
%store -r daily_deaths
%store -r date
%store -r days
```

```python
# cross-correlation usinf fft
def corr_fft(x, y, dt=1):
    
    

    # padding the arrays with N-1 zeros
    y = np.pad(y, (0,len(y)-1), mode='constant')
    x = np.pad(x, (0,len(x)-1), mode='constant')


    #fourier transforms
    FT_y = np.fft.fft(y)  # dt=1 so no need to multiply by dt
    FT_x = np.fft.fft(x)


    # multiply fourier transforms and then ifft to get the cross-correlation
    W = FT_x*np.conj(FT_y)  # notice that one is the complex conjugate of the fourier transform
    w = np.fft.fftshift(np.fft.ifft(W))
    
    # time lag axis
    tau = np.arange(0, len(x), 1) - len(x)//2

    return w, tau
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot_date(date[:-2], daily_cases[:-1], "r", label="New Cases")
plt.plot_date(date[:-2], daily_recoveries[:-1], "g", label="New  Recoveries")
ax.xaxis.set_minor_locator(days)
ax.xaxis.set_tick_params(rotation=30, labelsize=10)
#plt.yscale("log")
plt.title("Daily cases, recoveries, and deaths")
plt.legend(loc="upper left", framealpha=0, bbox_to_anchor=(-0.01,1))
plt.ylim([0,850])
plt.show()
```
